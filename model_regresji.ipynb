```python
#Wczytanie pliku csv
import pandas as pd

cleandata_df = pd.read_csv("clean_data.csv", sep=",")
```

```python
#Sprawdzenie czy dane są w porządku
cleandata_df.sample(5)
```

```python
#Usunięcie zbędnej kolumny przechowującej id
cleandata_df = cleandata_df.drop(cleandata_df.columns[0], axis=1)
```

```python
import statsmodels.formula.api as smf

myformula = "price ~ carat + C(clarity) + C(color) + C(cut) + x_dimension + y_dimension + z_dimension + depth + table"

model = smf.ols(formula=myformula, data=cleandata_df).fit()
print(model.summary())
```
